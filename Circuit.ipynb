```python
from qiskit import QuantumCircuit
import numpy as np

qc=QuantumCircuit(4)
#controlled Hadamard gate
qc.ch(0,1)
#controlled phase gate
qc.cp(np.pi/4,0,1)
#controlled RX gate
qc.crx(np.pi/4,1,2)
#controlled RY gate
qc.cry(np.pi/4,2,3)
#controlled RZ gate
qc.crz(np.pi/4,3,0)
#controlled swap gate
qc.cswap(0,1,2)
qc.csx(1,3)
#qc.cu(np.pi/4, np.pi/2, np.pi/3, 0,1)
qc.cx(0,1)
qc.cy(1,2)
qc.cz(2,3)
qc.dcx(0,1)
qc.iswap(1,2)
qc.mcp(np.pi/4, [0,1], 2)
qc.mcx([0,1],2)


qc.draw()
```

```python
from qiskit import QuantumCircuit
import numpy as np

def rotation(angle):
    qc=QuantumCircuit(2)
    qc.rz(angle,0)
    return qc.to_gate(label='alpha 01') #circuit name customize

qc=QuantumCircuit(3)
qc.h(0)
qc.cx(0,1)

my_gate=rotation(np.pi/2)
qc.append(my_gate,[0,2])
qc.draw()
```

```python
from qiskit import QuantumCircuit

qc = QuantumCircuit(2,2)
qc.h(0)
qc.cx(0,1)
qc.measure([0,1],[0,1])

inst = qc.to_instruction(label='alpha 2.0')

m_qc = QuantumCircuit(2,2)
m_qc.append(inst, [0,1], [0,1])

m_qc.draw()
```

```python

from qiskit import QuantumCircuit
import numpy as np
pi2=np.pi/2
qc=QuantumCircuit(4)
qc.rx(theta=pi2, qubit=0)
qc.cx(0,2)
qc.barrier()
qc.cx(1,3,ctrl_state=0)
qc.draw()
```

```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import numpy as np

qc = QuantumCircuit(3)

qc.ry(2*np.arccos(np.sqrt(2/3)), 0)
qc.cx(0,1)
qc.ry(2*np.arccos(1/np.sqrt(2)), 1)
qc.cx(1,2)

qc.measure_all()

sim = AerSimulator()
job = sim.run(qc, shots=1024)
result = job.result()

counts = result.get_counts()
print(counts)
qc.draw()
```
